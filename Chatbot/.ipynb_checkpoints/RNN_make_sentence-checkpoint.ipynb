```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
```

```python
df=pd.read_csv('ChatbotData.csv')
```

```python
df
```

```python
df.info()
```

```python
import re
```

```python
df['Q'] = df['Q'].apply(lambda x: re.sub(r'[^ ㄱ-ㅣ가-힣]+', '', x))
```

```python
df['Q'][:5]
```

```python
text='\n'.join(list(df['Q']))
text
```

```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts([text])
```

```python
len(tokenizer.word_index)+1, "단어 집합의 크기"
```

```python
tokenizer.word_index
```

```python
sequences=list()
for line in text.split('\n'):
    encoded=tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(encoded)):
        sequence=encoded[:i+1]
        sequences.append(sequence)
```

```python
len(sequences)
```

```python
sequences
```

```python
[k for k, v in tokenizer.word_index.items() if v == 4603]
```

```python
[k for k, v in tokenizer.word_index.items() if v == 346]
```

```python
[k for k, v in tokenizer.word_index.items() if v == 449]
```

```python
max(len(i) for i in sequences), "샘플의 최대 길이"
```

```python
sequences=pad_sequences(sequences, maxlen=15, padding='pre')
```

```python
sequences[0]
```

```python
sequences=np.array(sequences)
X=sequences[:,:-1]
y=sequences[:,-1]
```

```python
y=to_categorical(y, num_classes=13271)
```

```python
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, SimpleRNN
```

```python
embedding_dim=10
hidden_units=32

model=Sequential()
model.add(Embedding(13271, embedding_dim))
model.add(SimpleRNN(hidden_units))
model.add(Dense(13271, activation="softmax"))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y, epochs=10, verbose=2)
```
