```python
from surprise import SVD 
## 고유값분해(행렬을 더 낮은 차원으로 분해)해서 다시 원본으로 살리면서 잠재적인 값을 추론하는 방법
from surprise import Dataset, Reader ## SVD를 사용하기 위한 데이터셋 만들어주는 클래스
                            ## 사용자, 아이템, 평점
from surprise import accuracy ## RMSE, MAE, cross-validation(k-fold)
from surprise.model_selection import train_test_split ## 훈련/검증 데이터 분류
```

```python
## 1. 데이터셋을 만들어주자(사용자, 아이템, 평점),  훈련/검증 데이터 분류
data = Dataset.load_builtin('ml-100k') # 2세대(영화 cd를 배송주문) 업체 사이트에서 제공(kaggle)한 데이터셋
data
```

```python
trainset, testset = train_test_split(data, test_size= 0.25, random_state=0)
```

```python
testset[1] # 892번 사람이 291번 영화에 4.0 평점을 부여함
```

```python
testset[:5]
```

```python
## 2. SVD() 모델 선정
algo = SVD()
```

```python
## 3. 훈련용 데이터로 fit()
algo.fit(trainset)
```

```python
## 4. 검증용 데이터로 예측
prediction = algo.test(testset)[1]
prediction #하나의 영화에 대한 예측값은 Prediction객체!
```

```python
predictions = algo.test(testset)[:5]
predictions
# r_ui : 실제값, est : 예측값
```

```python
'userid: '+prediction.uid+', iid: '+prediction.iid+', est: '+str(prediction.est)
```

```python
predictions[0]
```

```python
predictions = algo.test(testset)
```

```python
for one in predictions:
    print('userid: '+one.uid+', iid: '+one.iid+', est: '+str(one.est))
```

```python
len(predictions)
```

```python
accuracy.rmse(predictions) #오차의 평균값
```

```python
from surprise.model_selection import cross_validate
```

```python
cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True) 
```

```python
action_movies=movies[movies['genres']=='Action|Crime|Drama'].reset_index()
```

```python
action_movies.loc[0,'movieId']
```

```python
uid='9'
max_iid='0'
max_est=0
for i in range(len(action_movies)):
    iid=str(action_movies.loc[i,'movieId'])
    
    if algo.predict(uid, iid, verbose=True).est > max_est:
        max_iid=algo.predict(uid, iid, verbose=True).iid
        max_est=algo.predict(uid, iid, verbose=True).est
print('max_iid >> ',max_iid)
print('max_est >> ',max_est)
```

```python
from surprise.dataset import DatasetAutoFolds
```

```python
reader = Reader(line_format='user item rating timestamp', sep=',', rating_scale=(0.5, 5))
# DatasetAutoFolds 클래스를 ratings_noh.csv 파일 기반으로 생성. 
data = DatasetAutoFolds(ratings_file='ratings_noh.csv', reader=reader)
```

```python
trainset, testset = train_test_split(data, test_size= 0.25, random_state=0)
```

```python
testset[0]
```

```python
algo.fit(trainset)
```

```python
algo.test(testset)[:3]
```

```python
accuracy.rmse(algo.test(testset))
```

```python
cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True) 
```
