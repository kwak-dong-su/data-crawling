```python
##1. 크롬 브라우저를 실행 
def connect():
    driver=webdriver.Chrome('C:/Users/hi/hi-AI/chromedriver')
    return driver
```

```python
driver_global=connect()
```

```python
## 2. 특정 페이지를 가지고 와서, 스크립트를 실행하자
def coffee_bean(no):
    driver_global.get('https://www.coffeebeankorea.com/store/store.asp')
    time.sleep(1)
    func="storePop2('"+str(no)+"')"
    shop=[]
    try:
        # 스크립트 실행
        driver_global.execute_script(func)
        time.sleep(1)
        # 실행한 결과 html 가지고 오세요
        html=driver_global.page_source
        # bsparser설정
        doc=BeautifulSoup(html, 'html.parser')
        
        #타이틀 검색
        title=doc.select_one('div.store_txt > h2').text
        #주소 검색
        addr=doc.select_one('.store_table > tbody > tr:nth-child(3) > td').text
        #전화번호 검색
        tel=doc.select_one('.store_table > tbody > tr:nth-child(4) > td').text
        
        shop=[title, addr, tel]
    except:
        print('해당번호없음')
    
    return shop #[삼성봉은사거리점 , 서울시 강남구... , 02-3443..]
```

```python
coffee_bean(6)
```

```python
title_list=[]
addr_list=[]
tel_list=[]
```

```python
for i in range(1, 200):
    print('---{0}번---'.format(i))
    shop=coffee_bean(i)
    if len(shop)!=0:
        title_list.append(shop[0])
        addr_list.append(shop[1])
        tel_list.append(shop[2])
        print(shop)
```

```python
df=pd.DataFrame({
                'title' : title_list,
                'addr' : addr_list,
                'tel' : tel_list
})
df
```

```python
df.to_csv('./coffee_bean.csv', index=False, encoding='utf-8-sig')
```
