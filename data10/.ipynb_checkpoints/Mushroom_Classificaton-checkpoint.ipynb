```python
j=0
dt_acc=[]
rf_acc=[]
svc_acc=[]
acc_list=[dt_acc, rf_acc, svc_acc]

for i, (train_index, test_index) in enumerate(kfold.split(mushrooms[t_data])):
    X_train2, X_test2=mushrooms[t_data].values[train_index], mushrooms[t_data].values[test_index]
    y_train2, y_test2=mushrooms['class'].values[train_index], mushrooms['class'].values[test_index]
    clf_name=['DT', 'RF', 'SVC']
    i=0
    
    for one in clf_list:
        one.fit(X_train2, y_train2)
        result=one.predict(X_test2)
        print(clf_name[i],':',accuracy_score(result, y_test2))    
        acc_list[i].append(accuracy_score(result, y_test2))
        i+=1
    print('------------fold '+str(j)+'------------')
    j+=1

print()
print('DT 평균: {0}'.format(np.mean(dt_acc)))
print('RF 평균: {0}'.format(np.mean(rf_acc)))
print('SVC 평균: {0}'.format(np.mean(svc_acc)))
acc={'DT':float(np.mean(dt_acc)), 'RF':float(np.mean(rf_acc)), 'SVC':float(np.mean(svc_acc))}

max_index=max(acc,key=acc.get)
print()
print('제일 정확도가 높은 모델은 {0}이고, 정확도는 {1}입니다.'.format(max_index, acc[max_index]))
```
