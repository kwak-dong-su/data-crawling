```python
import nltk
```

```python
str="청와대"
```

```python
f = open('reviews/txt/'+str+' 리뷰.txt','r', encoding='UTF-8')
review = f.read()
f.close()
```

```python
review
```

```python
tokens_review=t.nouns(review)
```

```python
tokens_review
```

```python
stopwords = [str, "또한", "방문", "여행","시간", "타고", "정말", "가장", "항상", "도착", "다른", "매우", "경우", "우리", "이용" ,"모든", "때문", "장소", "지역", "그것"]
```

```python
tokens_review[1]
```

```python
tokens_review2=[w for w in tokens_review if w not in stopwords and len(w)>1]
```

```python
tokens_review2
```

```python
fd_review=FreqDist(tokens_review2)
```

```python
wc = WordCloud(width=2000, height=1300, background_color="white", random_state=0, font_path=font_location)
plt.imshow(wc.generate_from_frequencies(fd_review))
plt.axis("off") # 축 제거
# plt.show()
plt.savefig("reviews/img/"+str+".png") # 이미지 저장하려면 plt.show() 주석처리해야함
```
