```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df
```

```python
df.dtypes
```

```python
df[df['age'] >= 30]['age'].value_counts().head()
```

```python
# 나이가 30살 이상 남자 승객 조건 필터링
# fare를 많이 낸 순서로 내림차순 정렬
# 상위 10개를 출력
```

```python
df[(df['age'] >= 30) & (df['sex']=='male')].head(10)
```

```python
df.sort_values(by='fare', ascending=False).head(10)
```

```python
# 나이가 20살 이상 40살 미만인 승객
# pclass가 1등급 혹은 2등급인 승객
# 열(column)은 survived, pclass, age, fare 만 나오게 출력
# 10개만 출력
```

```python
df[(df['age'] >= 20) & (df['age']<40)].loc[:,["survived","pclass", "age", "fare"]].head(10)
```

```python
df[(df['pclass'] == 1) | (df['pclass']==2)].loc[:,["survived","pclass", "age", "fare"]].head(10)
```
