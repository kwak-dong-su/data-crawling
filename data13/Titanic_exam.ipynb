```python
import pandas as pd
```

```python
titanic=pd.read_csv('../csv-data/titanic_train.csv')
```

```python
titanic
```

```python
# 1번
```

```python
titanic['pclass']=titanic['pclass'].apply(lambda x:'low' if x==1 else('middle' if x==2 or x==3 else 'high'))
titanic
```

```python
# 2번
```

```python
titanic['name_result']=titanic['name'].apply(lambda x:'long' if len(x)>=15 else 'short')
titanic
```

```python
# 3번
```

```python
titanic['survived']=titanic['survived'].apply(lambda x:'생존' if x==1 else '사망')
titanic
```

```python
# 4번
```

```python
titanic['age']=titanic['age'].apply(lambda x:'60대' if x>=60 else('50대' if x>=50 else('40대' if x>=40 else('30대' if x>=30 
                                                else('20대' if x>=20 else('10대' if x>=10 else '그 외'))))))
titanic
```
