```python
knn=KNeighborsClassifier()
```

```python
id(knn)
```

```python
knn.fit(X_train, y_train)
```

```python
# 샘플링을 10개해서, 10번 검증을 해보세요! 의미
# X_train을 train data, validation data로 나눈다.
# cv(fold값)가 10이면 validation data를 X_train의 1/10로 잡음
# validation data의 개수는 12개, 나머지가 train data

scores=cross_val_score(knn, X_train, y_train, cv=10) #cv=fold값!
scores
```

```python
pred=knn.predict(X_test)
acc=accuracy_score(y_test, pred)
acc
```

```python
def print_metrics(model, x, y, cv):
    scores=cross_val_score(model, x, y, cv=cv) #cv=fold값!
    print('---------------교차검증 결과-------------\n{0}'.format(scores))
    acc_mean=np.mean(scores)
    print('---------------교차검증 평균-------------\n{:.2f}'.format(acc_mean))
```

```python
print_metrics(knn, X_train, y_train, 10)
```

```python
## 스케일링한 데이터로 knn fit-> 교차검증과 test data로 검증해서 비교해보세요!
print_metrics(knn, X_test, y_test, 10)
```

```python
print_metrics(knn, iris_df.iloc[:,:-1], iris_df.iloc[:,-1], 10)
```

```python
## 대략적인 모델 2-3개를 선정했다고 가정
## 모델마다 최적의 하이퍼파라메터가 무엇인지 찾아주는 것이
## grid search
```

```python
grid_params={
    'n_neighbors':list(range(1, 20)), #19개
    'weights':['uniform', 'distance'], #2개
    'metric':['euclidean', 'manhattan', 'minkowski'] #3개
} ## 19 * 2 * 3
# fold= 5,  19 * 2 * 3 * 5
```
