```python
## 라벨 인코딩할 컬럼이 2~3개인 경우
```

```python
words = ['apple', 'summer', 'summer', 'spring', 'spring']
hobby = ['book', 'run','talk','coffee','song']
```

```python
df=pd.DataFrame({
    'words':words,
    'hobby':hobby
})
df
```

```python
encoder2=LabelEncoder()
```

```python
col=df.columns
col
```

```python
for one in col:
    df[one]=encoder2.fit_transform(df[one])
```

```python
df
```

```python
#나이(100, 200, 300) : 연속적인 숫자 => 라벨인코딩의 대상이 되지 않음.
#나이대(어린이, 성년, 노년) :연속적인 숫자값이 아님=>라벨인코딩의 대상이 됨. category
```

```python
## 원핫 인코딩! ###
### 라벨인코딩: 연속형이 아닌 데이터를 원하는 형태의 숫자로 바꾸어주는 것!
###              1)연속형이 아닌 데이터를 숫자로 변환: LabelEncodertt사용
###                fit()--transform(), fit_transform()
###              2) 숫자로만 변환하는 경우, 컴퓨터는 연속형 숫자로 오해하여 연산수행하여
###                 의도치않은 결과가 나올 수 있음. 해당됨을 나타나게 하기 위해
###                 라벨의 갯수만큼 컬럼을 만들어주고 해당되는 라벨에 표시(1)로 해주는 방법을 사용
###                 원핫인코딩
```

```python
#3개로 group. 라벨인코딩/원핫인코딩
#1. beer_servings를 3개의 그룹으로 나누기
#2. 1번의 결과를 bear_cat 변수로 추가
drink=pd.read_csv('drinks.csv')
drink
```

```python
drink['bear_cat']=pd.cut(
                            drink['beer_servings'],
                            bins=[0,100,200,400],
                            include_lowest=True,
                            labels=['few','medium','many']
                          )
drink[drink['bear_cat']=='few']
```

```python
drink.sort_values('beer_servings', ascending=False)
```

```python
#라벨 인코딩
```

```python
bear=['few', 'medium', 'many']
data=pd.Series(bear)
encoder=LabelEncoder()
encoder.fit(data)
```

```python
target=encoder.transform(drink['bear_cat'])
print(target, type(target), len(target))
```

```python
encoder.inverse_transform(target)
```
